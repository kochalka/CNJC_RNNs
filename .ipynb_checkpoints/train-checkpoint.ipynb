```python
import numpy as np
from scipy.io import loadmat
from rnn_utils import RateUnitNetwork
import torch
from torch import nn, optim
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
device = 'cuda:0'
input_dim = 2
hidden_dim = 800
output_dim = 2 
noise_std = 0.001
```

```python
model = RateUnitNetwork(input_dim,
                        hidden_dim,
                        output_dim,
                        include_bias = False,
                        dt=1,
                        tau=10,
                        device = device,
                        noise_amp=noise_std,
                       )

model = model.to(device)
```

```python
handwriting_data_mat = './nn.3405-S2/DAC_handwriting_training/DAC_handwriting_output_targets.mat'
handwriting_data = loadmat(handwriting_data_mat)
```

```python
num_epochs = 50000
learning_rate = 1e-3
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), 
                      lr=learning_rate, 
                      weight_decay=1e-4)
```

```python
stimulus_amp = 2.0 # Amplitude of the input pulse
t_init = 200 # Time to wait before the stimulation
t_stim= 50 # Duration of the stimulation
#t_relax = 150 # Duration to relax after the trajectory
```

```python
patterns = ['neuron', 'chaos']
targets = {}
inputs = {}
for j, pat in enumerate(patterns):
    target = handwriting_data[pat].astype(np.float32)
    targets[pat] = torch.from_numpy(target.T).to(device=device).view((1,) + target.T.shape)
    t_total = t_init + t_stim + target.shape[1]
    inputs[pat] = torch.zeros((1, t_total, 2), device=device)
    inputs[pat][0, t_init:t_init+t_stim, j] = stimulus_amp
```

```python
# i0 = 14300
# train_loss = np.r_[train_loss[:i0], np.zeros(50000-i0)]
#train_loss = np.r_[train_loss, np.zeros(15000)] #FIXME
```

```python
#model = model.to(device=device)
```

```python
batch_size = 10
train_loss = np.zeros(num_epochs)
for i in range(0,num_epochs):
    for pat in patterns:
        # Initialize hidden
        hidden = 2*torch.rand((batch_size, hidden_dim), device=device) - 1
        
        if batch_size>1:
            inp = inputs[pat].repeat(batch_size,1,1)
            trg = targets[pat].repeat(batch_size,1,1)
        else:
            inp = inputs[pat]
            trg = targets[pat]
            
        
        outputs, hiddens = model(inp, hidden)
        loss = criterion(outputs[:,(t_init+t_stim):], #-t_relax
                         trg)
        model.zero_grad()
        loss.backward()
        optimizer.step()
        train_loss[i]+=loss.data.item()
    if (i%25)==0:
        print('epoch: {}, train_loss = {}'.format(i, train_loss[i]))
    if (i%1000) == 0:
        torch.save(model.state_dict(), 
                   '/media/druckmannuser/4C82D91382D901FE/handwriting_models/handwriting_mdl_new2_{}.pt'.format(i))

    #TODO: TEST TBPTT. FORNOW: ignore and don't use truncation
    #if tbptt_window is not None: # Use TBPTT
    #    batch_inputs = torch.split(inputs, tbptt_window, dim = 1)
    #    batch_targets = torch.split(targets, tbptt_window, dim = 1)
    #    for inp, tgt in zip(batch_inputs, batch_targets):
    #        hidden = hiddens[:,-1,:].detach()
    #        outputs, hiddens = model(inp, hidden)
    #        loss = criterion(outputs.to(device=device), tgt)
    #        model.zero_grad()
    #        loss.backward()
    #        optimizer.step()
    #        train_loss[i]+=loss.data.item()
    #else: # Use BPTT on full sequence
"""
    if (i%1000) == 0:
        state = {
            'state_dict': model.state_dict(),
            'optimizer': optimizer.state_dict(),
            'input_domain' : args.input_domain,
            'pulse_width' : args.pulse_width,
            'pot_dims' : args.pot_dims,
            'null_dims' : null_dims,
            'input_delay' : args.input_delay,
            'num_steps_train' : args.num_steps_train
            }
        save_checkpoint(state, i)

"""
```

```python
device = 'cpu'
model=model.to(device=device)
model.device = device

for pat in patterns:
    fig, ax = plt.subplots(1,3,figsize=(12,3))
    for i in range(3):
        hidden = 2*torch.rand((batch_size, hidden_dim), device=device) - 1
        #outputs, hiddens = model(torch.zeros(1,2500,2), hidden.to(device))
        #foo = torch.zeros(1,2500,2)
        #foo[:,200:250,:] = 0.2
        #outputs, hiddens = model(foo, hidden.to(device))
        outputs, hiddens = model(inputs[pat].to(device=device), 
                                 hidden.to(device))

        trace_out = outputs.detach().cpu().numpy()[0].T
        trace_out = trace_out[:,(t_init+t_stim):] #-t_relax]
        ax[i].scatter(*trace_out, 
                      c=np.arange(trace_out.shape[1]))
        ax[i].scatter(*targets[pat].detach().cpu().numpy()[0].T, 
                      marker='x', s=0.1)
plt.show()
```
